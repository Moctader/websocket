```python
end_date = datetime.datetime.utcnow()
start_date = end_date - datetime.timedelta(days=5)
utc_end_date = calendar.timegm(end_date.utctimetuple())
utc_start_date = calendar.timegm(start_date.utctimetuple())
print("utc_start_date:", utc_start_date, "utc_end_date:", utc_end_date)
```

```python
forex_eur='EUR.FOREX'
usd_eur=client.get_prices_intraday('USD.FOREX', s=forex_eur, interval='5m', start=utc_start_date, end=utc_end_date)
usd_eur_df=pd.DataFrame(usd_eur)
usd_eur.set_index('datetime', inplace=True)
usd_eur.close.plot(figsize=(10,5))

plt.title('USD/EUR 5m time bars')
plt.ylabel('Price')
plt.show()
```
